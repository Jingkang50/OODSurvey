```python
import re

with open('AD.md', 'r') as reader, open('AD_reformat.md', 'w') as writer:
    for line in reader.readlines():
        if 'title=' in line and not 'booktitle=' in line:
            title = re.search('{(.*)}', line).group(1)
        if 'author=' in line:
            author = re.search('{(.*)}', line).group(1)
        if 'booktitle=' in line or 'journal=' in line:
            book = re.search('{(.*)}', line).group(1)
            if 'arXiv' in book:
                link = book.strip('arXiv preprint arXiv:')
                book = 'arXiv'
                link = 'https://arxiv.org/abs/'+link
            else:
                link = ''
        if 'year=' in line:
            year = re.search('{(.*)}', line).group(1)
            
        if line.startswith('}'):
            writer.writelines(f'**[{book}-{year}]**\n')
            writer.writelines(f'[{title}]({link})\n')
            writer.writelines('<br>\n')
            writer.writelines(f'**Authors:** {author}\n')
            writer.writelines('<br>\n')
            writer.writelines('**Institution:** \n')
            writer.writelines('> <details>\n')
            writer.writelines('> <summary></summary>\n')
            writer.writelines('> <p style="text-align:left">\n')
            writer.writelines('>\n')
            writer.writelines('> </p>\n')
            writer.writelines('> </details>\n')
            writer.writelines('\n')
        
        if line.startswith('%'):
            writer.writelines(line + '\n')
```
